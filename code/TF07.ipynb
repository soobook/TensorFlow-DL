## 패션 MNIST 데이터셋 앞부분 이미지 10개

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 패션 MNIST 데이터셋 로딩
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# 클래스 이름 정의
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# 1. 데이터의 크기 출력
print("훈련 데이터 수:", x_train.shape[0])
print("테스트 데이터 수:", x_test.shape[0])
print("이미지 크기:", x_train.shape[1:], "(픽셀)")
print("클래스 수:", len(np.unique(y_train)), "개")

# 2. 샘플 이미지 10개 출력
plt.figure(figsize=(6, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"{class_names[y_train[i]]}")
    plt.axis('off')
plt.suptitle("Fashion-MNIST 샘플 이미지 (10개)", fontsize=16)
plt.tight_layout()
plt.show()
```
