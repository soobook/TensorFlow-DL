## 패션 MNIST 데이터셋 준비와 이미지 데이터 시각화

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 패션 MNIST 데이터셋 로딩
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# 1. 데이터의 크기 출력
print("학습 데이터 shape:", x_train.shape)
print("학습 데이터 수:", x_train.shape[0])
print("테스트 데이터 수:", x_test.shape[0])
print("이미지 크기:", x_train.shape[1:], "(픽셀)")
print("클래스 수:", len(np.unique(y_train)), "개")
```

```python
# 2. 테스트 데이터 열 번째 이미지와 정답(라벨) 출력

# 분류 클래스 이름 정의
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
idx = 9

plt.figure(figsize=(3, 2))
plt.imshow(x_test[idx], cmap='gray')
plt.title(f"{class_names[y_test[idx]]} {y_test[idx]}")
plt.axis('off')
plt.show()
```

```python
# 3. 학습 데이터 이미지 10개 출력

# 분류 클래스 이름 정의
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(6, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"{class_names[y_train[i]]} {y_train[i]}", fontsize=11)
    plt.axis('off')
plt.suptitle("Fashion-MNIST 학습 데이터 이미지 (10개)", fontsize=16)
plt.tight_layout()
plt.show()
```
